## Introducción:

El objetivo de la asignación es realizar un análisis descriptivo de los datos temporales, geográficos y de ususarios del servicio de alquiler de bicicletas en la ciudad de New Jersey durante el año 2019.
Citi Bike es el programa de bicicletas compartidas más grande de Estados Unidos. Citi Bike se lanzó en mayo de 2013, es un sistema divertido, eficiente y asequible, sin mencionar que es saludable y bueno para el medio ambiente, está disponible para su uso las 24 horas del día, los 7 días de la semana, los 365 días del año. Citi Bike, al igual que otros sistemas de bicicletas compartidas, consta de una flota de bicicletas especialmente diseñadas, que pueden desbloquearse desde una estación y devolverse a cualquier otra estación del sistema, lo que las hace ideales para viajes de ida. Las personas usan bicicletas compartidas para ir al trabajo o a la escuela, hacer mandados, llegar a citas o compromisos sociales, y más.
El dataset está conformado por un archivo CSV, de 15 columnas y 404.750 registros, que contienen información del servicio de alquiler de bicicletas de Citi Bike durante el año 2019 en la ciudad de New Jersey. El mismo dispone de datos relacionados al alquiler de bicicletas de la ciudad de New Jersey durante el año 2019. Las bicicletas pueden ser rentadas por dos tipos de usuarios: Subscriber y Customer.

### 2.2 Top 10 recorridos/trayectos realizados

```python
n_by_start_station_id = pd.DataFrame(df_csv,columns=['tripnumber','start station name', 'end station name'])
n_by_start_station_id = n_by_start_station_id.groupby(["start station name", "end station name"])["tripnumber"].count()
n_by_start_station_id = n_by_start_station_id.to_frame()
n_by_start_station_id = n_by_start_station_id.sort_values(by = ['tripnumber'], ascending=False).head(10)
n_by_start_station_id = n_by_start_station_id.rename({'tripnumber': 'Numero_viajes'}, axis=1)
n_by_start_station_id.reset_index(inplace=True)
n_by_start_station_id["Rutas"] = n_by_start_station_id["start station name"].astype(str)+ ' --> ' + n_by_start_station_id["end station name"]
n_by_start_station_id.drop(columns=['start station name', 'end station name'])

sns.set_theme(style="whitegrid")
f, ax = plt.subplots(figsize=(7, 10))
ax = sns.barplot(y=n_by_start_station_id.Rutas,x=n_by_start_station_id.Numero_viajes,orient = 'h')
ax.bar_label(ax.containers[0])

plt.show()
```

### 2.3 Top 10 de las estaciones más usadas

```python
viajes_por_estaciones_salida = df_csv[['tripnumber','start station name']]
viajes_por_estaciones_salida = viajes_por_estaciones_salida.groupby(["start station name"])["tripnumber"].count()
viajes_por_estaciones_salida = viajes_por_estaciones_salida.to_frame()
viajes_por_estaciones_salida = viajes_por_estaciones_salida.sort_values(by = ['tripnumber'], ascending=False).head(10)
viajes_por_estaciones_salida = viajes_por_estaciones_salida.rename({'tripnumber': 'Salidas'}, axis=1)
viajes_por_estaciones_salida = viajes_por_estaciones_salida.rename_axis('Estacion')

viajes_por_estaciones_llegadas = df_csv[['tripnumber','end station name']]
viajes_por_estaciones_llegadas = viajes_por_estaciones_llegadas.groupby(["end station name"])["tripnumber"].count()
viajes_por_estaciones_llegadas = viajes_por_estaciones_llegadas.to_frame()
viajes_por_estaciones_llegadas = viajes_por_estaciones_llegadas.sort_values(by = ['tripnumber'], ascending=False).head(10)
viajes_por_estaciones_llegadas = viajes_por_estaciones_llegadas.rename({'tripnumber': 'Llegadas'}, axis=1)
viajes_por_estaciones_llegadas = viajes_por_estaciones_llegadas.rename_axis('Estacion')

estaciones_mas_usadas = pd.concat([viajes_por_estaciones_salida, viajes_por_estaciones_llegadas], axis=1)

X_axis = np.arange(len(estaciones_mas_usadas.index))

fig = plt.figure()

plt.bar(X_axis - 0.2, estaciones_mas_usadas.Salidas, 0.4, label = 'salidas')
plt.bar(X_axis + 0.2, estaciones_mas_usadas.Llegadas, 0.4, label = 'llegadas')


fig.set_figwidth(20)
fig.set_figheight(16)

plt.xticks(X_axis, estaciones_mas_usadas.index,rotation = 75)
plt.xlabel("Estaciones")
plt.ylabel("Numero de usuarios")
plt.title("Top estaciones mas usadas")
plt.legend()
plt.show()
```

### 3.2 Gráfico de torta: Distribución de usuarios por género

```python
prom_MF=df_csv.gender.value_counts()/df_csv["gender"].count()*100
values = [prom_MF[0],prom_MF[1],prom_MF[2]]
labels=["Desconocido", "Masculino", "Femenino"]
colors= ["#B2B2B2","#3968C5", "#FE5000"]
fig1, ax1 = plt.subplots(figsize=(12, 8))
ax1.pie(values, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Distribución de usuarios por género", fontsize=20, fontweight="bold", color='black')
plt.show()
```

La distribución porcentual por género se conforma de la siguiente forma: 6,5% desconocidos, 22,8% femenino y la mayoría 70,7% son de género masculino.

## 3.3 Gráfico de torta de tipo de usuario (customer y subscriber)

```python
prom_MF=df_csv.usertype.value_counts()/df_csv["usertype"].count()*100
users = [prom_MF["Subscriber"],prom_MF["Customer"]]
labels=["Subscriber", "Customer"]
colors= ["#3968C5", "#FE5000"]
fig2, ax2 = plt.subplots(figsize=(12, 8))
ax2.pie(users, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Distribución de usuarios por tipo de suscripción", fontsize=20, fontweight="bold", color='black')
plt.show()
```

La distribución porcentual por tipo de suscripción se conforma de la siguiente forma: 10,6% Customer, es decir, usuarios ocasionales que no pagan ningún tipo de mensualidad o suscripción y 89,4% usuarios que poseen una membresía o suscripción.

## 3.4 Gráfico de violín (género, edad, tipo de usuario)

```python
df_csv["gender2"]=df_csv["gender"].replace(0,"D")
df_csv["gender2"]=df_csv["gender2"].replace(1,"M")
df_csv["gender2"]=df_csv["gender2"].replace(2,"F")
df_csv
```

```python
colores = ["#3968C5", "#FE5000"]
ax = sns.catplot(data=df_csv, kind='violin', palette = colores, x='gender2', y='age', hue='usertype', split=True, height=8, aspect=15/8)
ax.set(xlabel='Tipo de Usuario', ylabel='Edad', title='Distribución del uso de bicicletas por edad, género y  tipo de suscripción')
```

En el gráfico de violin se muestra la distribución por edad de los tipos de usuario según el género
En el primer violín podemos deducir que los usuarios masculinos de tipo subscriber tienen una edad promedio de 31 años, en cambio en los usuarios masculinos de tipo customer tienen una edad promedio de 28 años.
En el segundo violín podemos deducir que las usuarias femeninas de tipo subscriber tienen una edad promedio de entre 30 y 35 años, en cambio en las usuarias femeninas de tipo customer tienen una edad promedio de 29 años.
Por último, en el tercer violín se hace notar que para los usuarios de tipo subscriber y customer la edad promedio es alrededor de 50 años.
