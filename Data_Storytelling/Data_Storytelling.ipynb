```python
import pandas as pd
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
import datetime as dt
from datetime import datetime
```

```python
url ='https://raw.githubusercontent.com/Andresmup/Proyecto_Data_Science_Bicicletas/master/Data_Wrangling/Dataset_2019.csv'
df_csv = pd.read_csv(url)
df_csv
```

```python
df_csv['tripnumber'] = df_csv.index
df_csv
```

```python
n_by_start_station_id = pd.DataFrame(df_csv,columns=['tripnumber','start station name', 'end station name'])
n_by_start_station_id = n_by_start_station_id.groupby(["start station name", "end station name"])["tripnumber"].count()
n_by_start_station_id = n_by_start_station_id.to_frame()
n_by_start_station_id = n_by_start_station_id.sort_values(by = ['tripnumber'], ascending=False).head(10)
n_by_start_station_id = n_by_start_station_id.rename({'tripnumber': 'Numero_viajes'}, axis=1)
n_by_start_station_id.reset_index(inplace=True)
n_by_start_station_id["Rutas"] = n_by_start_station_id["start station name"].astype(str)+ ' --> ' + n_by_start_station_id["end station name"]
n_by_start_station_id.drop(columns=['start station name', 'end station name'])

sns.set_theme(style="whitegrid")
f, ax = plt.subplots(figsize=(7, 10))
ax = sns.barplot(y=n_by_start_station_id.Rutas,x=n_by_start_station_id.Numero_viajes,orient = 'h')
ax.bar_label(ax.containers[0])

plt.show()
```

```python
viajes_por_estaciones_salida = df_csv[['tripnumber','start station name']]
viajes_por_estaciones_salida = viajes_por_estaciones_salida.groupby(["start station name"])["tripnumber"].count()
viajes_por_estaciones_salida = viajes_por_estaciones_salida.to_frame()
viajes_por_estaciones_salida = viajes_por_estaciones_salida.sort_values(by = ['tripnumber'], ascending=False).head(10)
viajes_por_estaciones_salida = viajes_por_estaciones_salida.rename({'tripnumber': 'Salidas'}, axis=1)
viajes_por_estaciones_salida = viajes_por_estaciones_salida.rename_axis('Estacion')

viajes_por_estaciones_llegadas = df_csv[['tripnumber','end station name']]
viajes_por_estaciones_llegadas = viajes_por_estaciones_llegadas.groupby(["end station name"])["tripnumber"].count()
viajes_por_estaciones_llegadas = viajes_por_estaciones_llegadas.to_frame()
viajes_por_estaciones_llegadas = viajes_por_estaciones_llegadas.sort_values(by = ['tripnumber'], ascending=False).head(10)
viajes_por_estaciones_llegadas = viajes_por_estaciones_llegadas.rename({'tripnumber': 'Llegadas'}, axis=1)
viajes_por_estaciones_llegadas = viajes_por_estaciones_llegadas.rename_axis('Estacion')

estaciones_mas_usadas = pd.concat([viajes_por_estaciones_salida, viajes_por_estaciones_llegadas], axis=1)

X_axis = np.arange(len(estaciones_mas_usadas.index))

fig = plt.figure()

plt.bar(X_axis - 0.2, estaciones_mas_usadas.Salidas, 0.4, label = 'salidas')
plt.bar(X_axis + 0.2, estaciones_mas_usadas.Llegadas, 0.4, label = 'llegadas')


fig.set_figwidth(20)
fig.set_figheight(16)

plt.xticks(X_axis, estaciones_mas_usadas.index,rotation = 75)
plt.xlabel("Estaciones")
plt.ylabel("Numero de usuarios")
plt.title("Top estaciones mas usadas")
plt.legend()
plt.show()
```
